# Put widgets including plot into a layout container

Source: https://stackoverflow.com/a/48349991

```python
from ipywidgets import *
import seaborn as sns
df = sns.load_dataset('iris')
import matplotlib.pyplot as plt

#plot
def plot_image(x):

    data = df

    if x != 'Select':
        xplot = data[x]
        sns.distplot(xplot)
        plt.show()

#define widgets
select_x = widgets.Dropdown(
        options=df.columns,
        value=df.columns[0],
        description='X'
    )

select_x.set_title  = 'x'

#assign widgets to tabs
tab_visualise = widgets.HBox([select_x])

#create tabs
tab_nest = widgets.Tab()
# tab_nest.children = [tab_visualise]
tab_nest.set_title(0, 'Visualise')


#interact function in isolation
f = interactive(plot_image, x = select_x);
tab_nest.children = [VBox(children = f.children)]
display(tab_nest)
```

## ugly matplotlib!

```python
from IPython.display import display
import matplotlib.pyplot as plt
plt.style.use('classic')
####%matplotlib inline 
####%matplotlib nbagg
%matplotlib widget
import numpy as np
import pandas as pd


# Create some data
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)


# Plot the data with Matplotlib defaults
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

## beautify matplotlib with seaborn `sns.set()`!

```python
import seaborn as sns
sns.set()

# same plotting code as above!
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');

```
